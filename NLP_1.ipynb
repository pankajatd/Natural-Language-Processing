Use Case : Ticket management System.

1.**Data Acquistion**(To get teh Training Data from the Database).

2.**Text Extraction** and  Cleanup(Removing Irrelavant Information,Spelling mistakes)

3.**PreProcessing: **
Step1: Sentence Tokenization(Breaking the sentence with Question mark,Full stop etc.NLTK and SpacY areused.)

Step2:Word Tokenization(Creating separate words).

Step3: Stemming(Getting the base word for each word. Example Eating===> Eat,Loves==>Love)
  Steeming does not get the base Word for each word..as it does not understand Grammar Example Ate===> Ate..it does not change it to Eat.
  To do we us Lemmatization.

  Examples of Stemming :

  Adjustable===>Adjust

  Formality===>Formali

  airliner===>airlin


Step4:Lemmatization(Getting the base word for each word by understanding the Grammar. Example Ate===> Eat)

Examples of Lemmatization :

Was ===> (to) be

better===> Good

meeting==>meeting

** These Steps that Step1 --Step4 is called as PrepRocessing**

4.**Feature Engineering** method is used to Convert Words to Number.(TF_IDF -Vectorizer,One-HotEncoding,WordEmbedding).

5.**Model Building**:Apply Machine Learning Classification Algorithm** To classify Tickets as High,Medium,Low Priority)
