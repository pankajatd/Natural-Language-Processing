**Tokenization in Spacy**

Tokenization ids the process of splitting the Text into meaningful **Segments**

Paragraph ====> Sentences ===> Words

```python
import spacy
```

```python
nlp=spacy.blank("en")
```

```python
doc=nlp("Dr. Strange loves pav bhaji of mumbai as it costsonly 2$ per plate")

for token in doc:
    print(token)
```

**Rules for Tokenization**


1.Split by prefix

"Let's go to N.Y!"

"--->split1 -->Let's go to N.Y!"

2.Split by Exception
Let--->split1 -->'s go to N.Y!"


3.Split by Postfix
Let--->split1 -->'s go to N.Y! Split-2--"

3.Split by Exception (!)
Let--->split1 -->'s go to N.Y ---Split3--! Split-2--"





```python
doc=nlp('''"Let's go to N.Y.!"''')

for token in doc:
    print(token)
```
