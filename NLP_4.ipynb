** Lemmitization and Stemming**


**Stemming** - Use fixed rules to remove able,ingto derive a Base Word.

talking--->talk
eating --->eat
adjustable--->adjust

Stemming may always not work ..
Eample: Eg: ability---> abil


**Lemmitization**- Use knowledge of the Language to derive a Base Word.

Ate---> eat
With Lemmitization it can be fixed:

Example:
ability--->ability



```python
import nltk
import spacy
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
words=['eating','eats','eat','ate','adjustable','rafting','ability','meeting']
for word in words:
       print(word,'|',stemmer.stem(word))
```

**LEMMITIZAtION**

```python
nlp=spacy.load('en_core_web_sm')
doc=nlp('eating eats eat ate adjustable rafting ability meeting better')
for token in doc:
  print(token,'|',token.lemma_,'|',token.lemma)
```

```python
doc=nlp("Mando talked for 3 hours although talking isn't his thing he became talkitive")
for token in doc:
  print(token,'|',token.lemma_,'|',token.lemma)
```

```python
# add an Attibute Ruler
ar=nlp.get_pipe('attribute_ruler')
ar.add([[{'Text':'Bro'}],[{'Text': 'Brah'}]],{'LEMMA':'Brother'})

doc=nlp("Bro,you wanna go? Brah,don't say no! I am exhausted ")
for token in doc:
  print(token.text,'|',token.lemma_)
```
