Dhaval ate Fruits

Dhaval==Noun

ate==Verb

Fruits--Noun


Noun--> person,Place Thing,Idea

Verb:Action or Experience.

Example:Elon flew to Mars yesterday.He carried biriyani masala with him

Elon==Noun

Mars ==Noun.
flew==Verb

He---Pronoun(substitue of a Noun)He ,she,they,them

Example:I ate many fruits

I: Pronoun

ate--Verb

fruits--Noun

many--Adjective(Adjective describes the Noun)

example: I slowly ate many fruits

slowly--Adverb(Adds meaning to a Verb)
